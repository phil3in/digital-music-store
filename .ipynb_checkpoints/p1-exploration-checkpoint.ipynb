### Organigramme Compagnie

Q01. Ecrivez et exécutez une unique requête listant chaque employée de la compagnie (Id, nom + prénom, Titre) ainsi que son manager direct (nom + prénom, Titre)

```python
# requête de départ
query1 = """
    SELECT e.EmployeeId AS 'Id employé',
           e.firstname || ' ' || e.lastname AS 'Employee',
           e.Title,
           m.EmployeeId AS 'Id manager',
           m.firstname || ' ' || m.lastname AS 'Manager',
           m.Title
    FROM employees e
    LEFT JOIN employees m ON e.ReportsTo = m.EmployeeId;
"""
```

```python
results = pd.read_sql_query(query1, conn)
results
```

Q02. Pour chaque artiste présent dans cette BDD, indiquez le nombre d'albums de cet artiste également répertorié (Liste triée par nom d'artiste)


```python
query2 = """
    SELECT artists.name AS "Nom de l'artiste", 
           COUNT(albums.AlbumId) AS "Nombre d'albums"
    FROM artists
    INNER JOIN albums ON artists.ArtistId = albums.ArtistId
    GROUP BY albums.ArtistId
    ORDER BY artists.name;
"""

results = pd.read_sql_query(query2, conn)
results
```

Q03. Produire une liste des albums reprenant également l'artiste de l'album, le nombre de pistes, la durée de toutes les pistes (en minutes) et le coût total (Liste triée par artiste puis nom d'album)


```python
query3 = """
    SELECT albums.Title AS "Album",
           artists.name AS "Nom de l'artiste", 
           COUNT(tracks.AlbumId) AS "Nombre de pistes",
           SUM(tracks.Milliseconds)/60000 AS "Durée en minutes",
           SUM(tracks.UnitPrice) AS "Coût total"
    FROM albums
    INNER JOIN artists ON albums.ArtistId = artists.ArtistId
    INNER JOIN tracks ON albums.AlbumId = tracks.AlbumId
    GROUP BY albums.AlbumId
    ORDER BY artists.name, albums.Title;
"""

results = pd.read_sql_query(query3, conn)
results
```

Q04. Produire une liste des toutes les playlists reprenant son nom, le nombre de pistes, la durée de toutes les pistes (en minutes) et le coût total (Liste triée par nom de playlist)


```python
query4 = """
    SELECT playlists.Name AS "Playlist",
           COUNT(playlist_track.TrackId) AS "Nombre de pistes",
           COALESCE(SUM(tracks.Milliseconds)/60000, 0) AS "Durée en minutes",
           COALESCE(SUM(tracks.UnitPrice), 0) AS "Coût total"
    FROM playlists
    -- Il y a des playlists vides, sans piste, pour les afficher quand même on utilise LEFT JOIN
    LEFT JOIN playlist_track ON playlists.PlaylistId = playlist_track.PlaylistId
    LEFT JOIN tracks ON playlist_track.TrackId = tracks.TrackId
    GROUP BY playlists.PlaylistId
    ORDER BY playlists.Name;
"""

results = pd.read_sql_query(query4, conn)
results
```

Q05. Produire une liste des tous les genres de musique reprenant son genre, le nombre de pistes, la durée de toutes les pistes (en minutes), le coût total et le coût moyen (Liste triée par nom de genre)


```python
query5 = """
    SELECT genres.Name AS "Genre",
           COUNT(tracks.TrackId) AS "Nombre de pistes",
           SUM(tracks.Milliseconds)/60000 AS "Durée en minutes",
           SUM(tracks.UnitPrice) AS "Coût total",
           AVG(tracks.UnitPrice) AS "Coût moyen"
    FROM genres
    JOIN tracks ON genres.GenreId = tracks.GenreId
    GROUP BY genres.GenreId
    ORDER BY Genres.Name;
"""

results = pd.read_sql_query(query5, conn)
results
```

Q06. Indiquez les caractéristiques principales de toutes les pistes correspondant au genre 'Science Fiction' (nom de piste, album, artiste de l'album, compositeur, nom du type de media, durée (en secondes) et prix unitaire


```python
query6 = """
    SELECT tracks.Name AS "Piste",
           albums.Title AS "Album",
           artists.Name AS "Artiste",
           tracks.Composer AS "Compositeur",
           media_types.Name AS "Type de média",
           (tracks.Milliseconds / 1000) AS "Durée en secondes",
           tracks.UnitPrice AS "Prix unitaire"
    FROM tracks
    LEFT JOIN albums ON tracks.AlbumId = albums.AlbumId
    LEFT JOIN artists ON albums.ArtistId = artists.ArtistId
    LEFT JOIN media_types ON tracks.MediaTypeId = media_types.MediaTypeId
    JOIN genres ON tracks.GenreId = genres.GenreId
    WHERE genres.Name = "Science Fiction"
    ORDER BY tracks.Name;
"""

results = pd.read_sql_query(query6, conn)
results
```

Q07. Retrouvez tous les clients (nom, prénom, email, pays) ayant acheté de la musique de type 'World'


```python
query7 = """
    SELECT DISTINCT customers.LastName AS "Nom",
           customers.FirstName AS "Prénom",
           customers.Email AS "Courriel",
           customers.Country AS "Pays"
    FROM customers
    JOIN invoices ON customers.CustomerId = invoices.CustomerId
    JOIN invoice_items ON invoices.InvoiceId = invoice_items.InvoiceId
    JOIN tracks ON invoice_items.TrackId = tracks.TrackId
    JOIN genres ON tracks.GenreId = genres.GenreId
    WHERE genres.Name = "World"
    ORDER BY customers.LastName;
"""

results = pd.read_sql_query(query7, conn)
results
```

Q08. Nommez les 15 artistes comptant le plus de pistes de type 'Rock'


```python
query7 = """
    SELECT artists.Name AS "Artiste",
           COUNT(tracks.TrackId) AS "Nombre de pistes"
    FROM artists
    JOIN albums ON artists.ArtistId = albums.ArtistId
    JOIN tracks ON albums.AlbumId = tracks.AlbumId
    JOIN genres ON tracks.GenreId = genres.GenreId
    WHERE genres.Name = "Rock"
    GROUP BY artists.Name
    ORDER BY COUNT(tracks.TrackId) DESC
    LIMIT 15;
"""

results = pd.read_sql_query(query7, conn)
results
```

```python
x = results["Artiste"]
y = results["Nombre de pistes"]

plt.figure()
plt.barh(x, y)
plt.title("Nombre de pistes listées par artiste")
plt.xlabel("Nombre de pistes")
plt.show()
```

Q09. Nommez les 5 meilleurs clients 

```python
query9 = """
    SELECT customers.LastName AS "Nom",
           customers.FirstName AS "Prénom",
           SUM(invoices.Total) AS "Dépenses totales"
    FROM customers
    JOIN invoices ON customers.CustomerId = invoices.CustomerId
    GROUP BY customers.CustomerId
    ORDER BY SUM(invoices.Total) DESC
    LIMIT 5;
"""

results = pd.read_sql_query(query9, conn)
results
```

Q10. Détaillez (piste, album, artiste, genre et type media) tous les achats effectués par le client 'Richard Cunningham' (triée par date achat)


```python
query10 = """
    SELECT invoices.InvoiceDate AS "Date",
           tracks.Name AS "Piste",
           albums.Title AS "Album",
           artists.Name AS "Artiste",
           genres.Name AS "Genre",
           media_types.Name AS "Type média"
    FROM customers
    JOIN invoices ON customers.CustomerId = invoices.CustomerId
    JOIN invoice_items ON invoices.InvoiceId = invoice_items.InvoiceId
    JOIN tracks ON invoice_items.TrackId = tracks.TrackId
    LEFT JOIN albums ON tracks.AlbumId = albums.AlbumId
    LEFT JOIN artists ON albums.ArtistId = artists.ArtistId
    LEFT JOIN genres ON tracks.GenreId = genres.GenreId
    LEFT JOIN media_types ON tracks.MediaTypeId = media_types.MediaTypeId
    WHERE customers.LastName = "Cunningham" AND customers.FirstName = "Richard"
    ORDER BY invoices.InvoiceDate;
"""

results = pd.read_sql_query(query10, conn)
results
```

Q11. Produire la liste de tous les contacts de la base de données (clients & employés) avec leur nom, prénom, type (client ou employé), email, pays et no de téléphone


```python
query11 = """
    SELECT customers.LastName AS "Nom",
           customers.FirstName AS "Prénom",
           "Client" AS "Type",
           customers.Email AS "Courriel",
           customers.Country AS "Pays",
           customers.Phone AS "N° de Téléphone"
    FROM customers
    UNION
    SELECT employees.LastName AS "Nom",
           employees.FirstName AS "Prénom",
           "Employé" AS "Type",
           employees.Email AS "Courriel",
           employees.Country AS "Pays",
           employees.Phone AS "N° de Téléphone"
    FROM employees
    ORDER BY LastName, FirstName;
"""

results = pd.read_sql_query(query11, conn)
results
```

Q12. Produire la liste de répartition des ventes par pays (basée sur `BillingCountry`)

```python
query12 = """
    SELECT invoices.BillingCountry AS "Pays",
           SUM(invoices.Total) AS "Ventes",
           (SUM(invoices.Total) / (SELECT SUM(invoices.Total) FROM invoices))*100 AS "Part en %"
    FROM invoices
    GROUP BY invoices.BillingCountry
    ORDER BY SUM(invoices.Total) DESC;
"""

results = pd.read_sql_query(query12, conn)
results
```

```python
x = results["Pays"]
y = results["Ventes"]
```

```python
plt.figure(figsize=(8, 6))
plt.barh(x, y)
plt.show()
```
