# P2. Les vues SQL

## Définition d'une vue SQL  

Une vue n'est rien de plus qu'une instruction SQLite qui est stockée dans la base de données avec un nom associé. Il s'agit en fait d'une composition d'une table sous la forme d'une requête SQLite prédéfinie.

Une vue peut contenir toutes les lignes d'une table ou des lignes sélectionnées d'une ou plusieurs tables. Une vue peut être créée à partir d'une ou plusieurs tables qui dépendent de la requête SQLite écrite pour créer une vue.

## Code utilitaire

```python
import os
import sqlite3
import pandas as pd   
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pathlib import Path
```

```python
# Ouvre connection vers SQLite db
chemin_bdd = Path('db') / 'chinook.db'
conn = sqlite3.connect(chemin_bdd)
```

```python
# Fonction permettant d'exécuter un requête SQL sur une BDD définie par sa connexion conn
def executer_requete(requete_sql, conn):
    try:
        cursor = conn.cursor()
        cursor.execute(requete_sql)
        conn.commit()
    except sqlite3.Error as e:
        print("Erreur lors de l'execution de la requête")
        print(e)
        return
    cursor.close()
```

## Example de vue SQL: v_albums

La requête suivante permet d'obtenir une liste (simplifiée) des albums / artistes correspondants:  

```python
# Requete sur les tables albums & artists
requete = """
    SELECT AlbumId,
           Title,
           a.Name AS Artist
    FROM albums
    JOIN artists a on albums.ArtistId = a.ArtistId
    ORDER BY Artist;
"""

albums = pd.read_sql_query(requete, conn)
albums.head(5)
```

Le code ci-dessous permet de créer une vue correspondant à la requête ci-dessous :    
Référence vue sous SQLite : https://www.sqlite.org/lang_createview.html

```python
# creation vue albums avec artistes
creation_vue_albums = """
    CREATE VIEW v_albums (
        AlbumId,
        Title,
        Artist)
    AS
        SELECT AlbumId,
               Title,
               a.Name AS Artist
        FROM albums
        JOIN artists a on albums.ArtistId = a.ArtistId
        ORDER BY Artist, Title;
"""

executer_requete(creation_vue_albums, conn)
```

Une fois la vue créé, on peut la réutiliser dans des requêtes SQL comme une table:  

```python
# Utilisation de la vue v_albums
albums = pd.read_sql_query("SELECT * FROM v_albums LIMIT 10;", conn)
albums.head(5)
```

```python
# Utilisation de la vue v_albums avec une clause WHERE
albums = pd.read_sql_query("SELECT * FROM v_albums WHERE Artist = 'U2';", conn)
albums
```

```python
# Utilisation de la vue v_albums avec une autre jointure
requete = """
    SELECT a.AlbumId,
           a.Title,
           a.Artist,
           count(t.TrackId) as Tracks
    FROM v_albums a
    JOIN tracks t ON a.AlbumId = t.AlbumId
    GROUP BY 1, 2, 3
    ORDER BY 3, 2;
"""
albums = pd.read_sql_query(requete, conn)
albums.head()
```

```python
# Effacer une vue
executer_requete("DROP VIEW IF EXISTS v_albums;", conn)
```

### Question: Avantages & Inconvénients des vues SQL

Quelles sont les avantages et inconvénients des vues SQL ?

Les vues permettent d'éviter d'avoir à constamment retaper les mêmes requêtes, ce qui fait économiser beaucoup de temps et de code. Elles permettent aussi d'utiliser une organisation des données différente et plus utile à la tâche qu'on désire accomplir, sans avoir à changer quoi que ce soit à l'organisation physique réelle de la base de données.

Mais côté machine, la requête n'est pas plus simple, au contraire, la formulation masque des requêtes imbriquées souvent inutilement compliquées ; si on abuse de ce système on en vient à revenir de nombreuses fois sur les mêmes tables, et on perd en performance.

### Vue 1: caractéristiques complètes des Albums  
A partir de l'exemple de vue ci-dessus, créer et exploiter une nouvelle vue v_albums reprenant pour chaque album:  
- AlbumId
- Title
- Artist
- nb de pistes
- Durée totale de l'album (en minutes)
- Prix de l'album

```python
# creation vue albums avec artistes
creation_vue_albums = """
    CREATE VIEW v_albums_plus (
        AlbumId,
        Title,
        Artist,
        TrackNb,
        Minutes,
        Price)
    AS
        SELECT albums.AlbumId,
               albums.Title,
               artists.Name AS "Artist",
               COUNT(tracks.TrackId) AS "TrackNb",
               SUM(tracks.Milliseconds)/60000 AS "Minutes",
               SUM(tracks.UnitPrice) AS "Price"
        FROM albums
        JOIN artists ON albums.ArtistId = artists.ArtistId
        JOIN tracks ON albums.AlbumId = tracks.AlbumId
        GROUP BY albums.AlbumId
        ORDER BY Artist, Title;
"""

executer_requete(creation_vue_albums, conn)
```

```python
# Utilisation de la vue v_albums
albums = pd.read_sql_query("SELECT * FROM v_albums_plus LIMIT 10;", conn)
albums.head(5)
```

### Vue 2: caractéristiques détaillées des pistes
Créer et exploiter une nouvelle vue v_tracks reprenant pour chaque piste:  
- TrackId
- Name
- Genre
- Album
- Artist
- Composer
- Temps (en seconde)
- MediaType

```python
# creation vue albums avec artistes
creation_vue_tracks = """
    CREATE VIEW v_tracks (
        TrackId,
        Name,
        Genre,
        Album,
        Artist,
        Composer,
        Seconds,
        MediaType)
    AS
        SELECT tracks.TrackId AS "TrackId",
               tracks.Name AS "Name",
               genres.Name AS "Genre",
               albums.Title AS "Album",
               artists.Name AS "Artist",
               tracks.Composer AS "Composer",
               (tracks.Milliseconds)/1000 AS "Seconds",
               media_types.Name AS "MediaType"
        FROM tracks
        LEFT JOIN albums ON tracks.AlbumId = albums.AlbumId
        LEFT JOIN artists ON albums.ArtistId = artists.ArtistId
        LEFT JOIN genres ON tracks.GenreId = genres.GenreId
        LEFT JOIN media_types ON tracks.MediaTypeId = media_types.MediaTypeId
        GROUP BY albums.AlbumId
        ORDER BY tracks.Name;
"""

executer_requete(creation_vue_tracks, conn)
```

```python
# Utilisation de la vue v_albums
albums = pd.read_sql_query("SELECT * FROM v_tracks LIMIT 10;", conn)
albums.head(5)
```

### Vue 3: Vue augmentée invoices
A. Créer une nouvelle vue v_invoices à partir de la table invoices ajoutant 3 nouveaux attributs basés sur l'attribut `InvoiceDate`:  
- semaine de la transaction
- mois de la transaction
- année de la transaction

```python
# creation vue invoices avec artistes
creation_vue_invoice_plus = """
    CREATE VIEW v_invoice_plus (
        InvoiceId,
        CustomerId,
        InvoiceDate,
        InvoiceWeek,
        InvoiceMonth,
        InvoiceYear,
        BillingAddress,
        BillingCity,
        BillingState,
        BillingCountry,
        BillingPostalCode,
        Total)
    AS
        SELECT i.InvoiceId,
               i.CustomerId,
               i.InvoiceDate,
               -- On rajoute 1 au numéro de semaine car sinon le décompte commence à 0 --
               STRFTIME("%W", i.InvoiceDate)+1 AS "InvoiceWeek",
               STRFTIME("%m", i.InvoiceDate) AS "InvoiceMonth",
               STRFTIME("%Y", i.InvoiceDate) AS "InvoiceYear",
               i.BillingAddress,
               i.BillingCity,
               i.BillingState,
               i.BillingCountry,
               i.BillingPostalCode,
               i.Total
        FROM invoices i;
"""

executer_requete(creation_vue_invoice_plus, conn)
```

```python
# Utilisation de la vue v_albums
albums = pd.read_sql_query("SELECT * FROM v_invoice_plus LIMIT 5;", conn)
albums.head(5)
```

B. Exploiter la vue v_invoice_plus ainsi crée pour déterminer les ventes annuelles par pays.

```python
queryB = """
    SELECT DISTINCT v.BillingCountry AS "Pays",
           nine.Ventes AS "Ventes 2009",
           ten.Ventes AS "Ventes 2010",
           eleven.Ventes AS "Ventes 2011",
           twelve.Ventes AS "Ventes 2012",
           thirteen.Ventes AS "Ventes 2013"
           
    FROM
    v_invoice_plus v
    
    LEFT JOIN
    (SELECT v.BillingCountry AS "Pays",
           SUM(Total) AS "Ventes" 
    FROM v_invoice_plus v
    WHERE InvoiceYear = "2009"
    GROUP BY v.BillingCountry) AS nine
    ON v.BillingCountry = nine.Pays
    
    LEFT JOIN
    (SELECT v.BillingCountry AS "Pays",
           SUM(Total) AS "Ventes" 
    FROM v_invoice_plus v
    WHERE InvoiceYear = "2010"
    GROUP BY v.BillingCountry) AS ten
    ON nine.Pays = ten.Pays
    
    LEFT JOIN
    (SELECT v.BillingCountry AS "Pays",
           SUM(Total) AS "Ventes"  
    FROM v_invoice_plus v
    WHERE InvoiceYear = "2011"
    GROUP BY v.BillingCountry) AS eleven
    ON nine.Pays = eleven.Pays
    
    LEFT JOIN
    (SELECT v.BillingCountry AS "Pays",
           SUM(Total) AS "Ventes" 
    FROM v_invoice_plus v
    WHERE InvoiceYear = "2012"
    GROUP BY v.BillingCountry) AS twelve
    ON nine.Pays = twelve.Pays
    
    LEFT JOIN
    (SELECT v.BillingCountry AS "Pays",
           SUM(Total) AS "Ventes" 
    FROM v_invoice_plus v
    WHERE InvoiceYear = "2013"
    GROUP BY v.BillingCountry
    ) AS thirteen
    ON nine.Pays = thirteen.Pays
    ;
"""

results = pd.read_sql_query(queryB, conn)
results
```

### Vue 4: A vous de jouer
Créer et exploiter une nouvelle vue de votre choix  

```python
creation_vue_artistos = """
    CREATE VIEW v_artistos (
        InvoiceId,
        InvoiceDate,
        Total,
        LastName,
        FirstName, 
        Country,
        Artiste,
        Genre)
    
    AS
        
        SELECT invoices.InvoiceId, 
               invoices.InvoiceDate,
               invoices.Total,
               customers.LastName, 
               customers.FirstName, 
               customers.Country, 
               artists.Name AS "Artiste", 
               genres.Name AS "Genre"
        FROM artists
        JOIN albums ON artists.ArtistId = albums.ArtistId
        JOIN tracks ON albums.AlbumId = tracks.AlbumId
        JOIN invoice_items ON tracks.TrackId = invoice_items.TrackId
        JOIN invoices ON invoice_items.InvoiceId = invoices.InvoiceId
        JOIN customers ON invoices.CustomerId = customers.CustomerId
        LEFT JOIN genres ON tracks.GenreId = genres.GenreId;
"""

executer_requete(creation_vue_artistos, conn)
```

```python
queryC = """
    SELECT v.Artiste, SUM(v.Total) AS "Ventes"
    FROM v_artistos v
    GROUP BY v.Artiste
    ORDER BY SUM(v.Total) DESC
    LIMIT 7
    ;
"""

results = pd.read_sql_query(queryC, conn)
results
```

```python
x = results["Artiste"]
y = results["Ventes"]

plt.figure(figsize=(8, 6))
plt.bar(x, y)
plt.title("Total des ventes par artiste (en $)")
plt.show()
```
